Task 1: Data Cleaning and Preprocessing <br>
Objective: Clean and prepare a raw dataset (with nulls, duplicates, inconsistent formats).<br>
Tools: Excel / Python (Pandas)<br>
Deliverables: Cleaned dataset + short summary of changes<br>



```python
#Importing libraries
import pandas as pd
import numpy as np
```

```python
#Reading the CSV file and seeing the top 5 rows from all columns
df=pd.read_csv(r"C:\Users\lnrka\OneDrive\Desktop\ELEVATE LABS\TASK_1\DATASETS\KaggleV2-May-2016.csv")
print(df.head())
```

```python
#By df.shape seeing the total number of rows and columns
print(df.shape)
```

```python
#By df.info() checking there are nulls are present and datatypes of each column
print(df.info())
```

```python
#Finding is there any missing values present
print("Missing Values : " )
print(df.isnull().sum())
```

```python
#Checking is there any duplicate rows
print("Duplicate Rows :")
print(df.duplicated().sum())
```

```python
#Changing the column names to lower and removing the spaces and speciak symbols and adding underscores(_)
df.columns = ['patient_id', 'appointment_id', 'gender', 'scheduled_day',
              'appointment_day', 'age', 'neighbourhood', 'scholarship',
              'hypertension', 'diabetes', 'alcoholism', 'handicap',
              'sms_received', 'no_show']

print(df.columns)
```

```python
#Changing the datatype of columns from str to datetime
df['scheduled_day']=pd.to_datetime(df['scheduled_day'])
df['appointment_day']=pd.to_datetime(df['appointment_day'])
print(df[['scheduled_day', 'appointment_day']].dtypes)
print(df[['patient_id', 'appointment_id']].head())
```

```python
#Checking is there any age in negative and if the min shows negative value then there is a negative value in it
print("Age Stats:")
print(df['age'].describe())
print("\nNegative ages:",(df['age']<0).sum())
```

```python
#Now filtering the age>=0 alone and removing the negative one
df=df[df['age']>=0]
print("After removing the invalid age:",df.shape[0])
```

```python
#Checking the values in Gender and no_show is there any unique values
print("Gender values: ",df['gender'].unique())
print("No_show values: ",df['no_show'].unique())
```

```python
#Changing the values all to capital to main a format
df['gender']=df['gender'].str.upper()
df['no_show']=df['no_show'].str.upper()
print("Corrected Gender Values",df['gender'].unique())
print("Corrected no_show values",df['no_show'].unique())
```

```python
#After cleaning, the new cleaned csv file is exported and saved
df.to_csv(r"C:\Users\lnrka\OneDrive\Desktop\ELEVATE LABS\TASK_1\DATASETS\cleaned_medical_appointments.csv", index=False)
print("Cleaned dataset saved successfully!")
```

```python
# Summary of changes
print("--- DATA CLEANING SUMMARY ---")
print(f"Total rows after cleaning: {df.shape[0]}")
print(f"Total columns: {df.shape[1]}")
print(f"\n1. Renamed all columns to lowercase with underscores")
print(f"2. Converted scheduled_day and appointment_day to datetime format")
print(f"3. Removed 1 row with invalid age (-1)")
print(f"4. Standardized gender values to uppercase (F, M)")
print(f"5. Standardized no_show values to uppercase (NO, YES)")
print(f"6. No missing values found")
print(f"7. No duplicate rows found")
print(f"\nCleaned dataset saved as cleaned_medical_appointments.csv")
```
